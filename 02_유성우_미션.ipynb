## 최적의 유성우 관측일

```python
# 도시명을 입력하면 관측가능한 유성우에 대한 정보를 알려주는 함수 작성하기

def predict_best_meteor_shower_viewing(city):
    # 안내 메시지 초기화
    meteor_shower_string = ''

    # 입력한 도시 정보가 존재하지 않으면 오류 메시지
    if city not in cities['city'].values:
        return f'{city}에서는 별자리 관측이 불가합니다.'
    
    # 도시의 위도 정보를 불러오기
    latitude = cities.loc[cities['city']==city, 'latitude'].iloc[0]
    
    # 해당 도시의 위도에서 관측 가능한 별자리 조회
    c_list = constellations.loc[(latitude<=constellations['latitudestart'])&(latitude >= constellations['latitudeend']), 'constellation'].tolist()

    # 볼 수 있는 별자리가 존재하지 않으면 오류 메시지
    if not c_list:
        return '볼 수 있는 별자리가 존재하지 않습니다.'
    
    # 별자리 관측 가능 안내 메시지
    print(f'{city}에서는 별자리 관측이 가능합니다:\n\n')
        
    
    # 도시별 관찰 가능한 별자리 iterate
    for c in c_list:
        # 별자리와 가장 가까운 유성우 조회
        ms = meteor_showers.loc[meteor_showers['radiant']==c, 'name'].iloc[0]
        
        # 유성우 관측이 가능한 시작일과 종료일
        meteor_showers_start = meteor_showers.loc[meteor_showers['radiant']==c, 'startdate'].iloc[0]
        meteor_showers_end = meteor_showers.loc[meteor_showers['radiant']==c, 'enddate'].iloc[0]
        
        # 유성우가 보일 때의 달의 위상 조회
        moon_phase_list = moon_phases.loc[(moon_phases['date'] >= meteor_showers_start) & (moon_phases['date'] <= meteor_showers_end)]

        # 달이 눈에 보이는 첫 날 조회
        b = moon_phase_list.loc[moon_phase_list['percentage'].idxmin()]['date']
        best_moon_date = b.to_pydatetime().strftime('%Y년 %m월 %d일')

        # 사용자에게 정보전달
        meteor_shower_string += f"[{ms}]를 잘 보시려면 [{c}]자리 위치를 향하여 [{best_moon_date}]에 보면 됩니다.\n"

    return meteor_shower_string
```
